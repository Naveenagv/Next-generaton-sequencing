```python
def correct_pvalues_for_multiple_testing(pvalues, correction_type = "Bonferroni"): 
    from numpy import array, empty                                                                        
    pvalues = array(pvalues) 
    n = pvalues.shape[0]
    print (n)
    new_pvalues = empty(n)
    if correction_type == "Bonferroni":                                                                   
        new_pvalues = n * pvalues
    return new_pvalues
```
